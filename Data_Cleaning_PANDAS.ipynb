# Pandas - Data Cleaning

```python
import pandas as pd
```

```python
df=pd.read_excel("D:\Customer Call List.xlsx")
df
```

```python
# Step 1. Drop Duplicates
df.drop_duplicates()
```

```python
# It has to be saved : assign a variable or use inplace( not a good practice )
df.count()
```

```python
df=df.drop_duplicates()
df
```

```python
df.count()
```

```python
# Step 2. Remove unwanted Columns
df=df.drop(columns='Not_Useful_Column')
df
```

```python
df
```

```python
# Step 3. Get rid of /,null, and other debris

# Be careful while saving this because DF!= 2 columns that has been stripped , the column = stripped form
#df['Last_Name']=df['Last_Name'].str.strip(['/','...']) -- this code gave nan values

df['Last_Name']=df['Last_Name'].str.strip('...')
df['Last_Name']=df['Last_Name'].str.strip('/')
df['Last_Name']=df['Last_Name'].str.strip('_')
df
```

```python
# Alter Method : one Go
df['Last_Name']=df['Last_Name'].str.strip('_,/_')
```

```python
df['Last_Name']
```

```python
# Step Working with Phone nos:

df['Phone_Number']=df['Phone_Number']=df['Phone_Number'].str.replace('[^a-zA-Z0-9]','',regex=True)
```

```python
df['Phone_Number']
```

```python
#Since we have strings & nos we convert it into strings
df['Phone_Number']=df['Phone_Number'].apply(lambda x:str(x))
df['Phone_Number']
```

```python
# Apply Lambda function for formatting

df['Phone_Number']=df['Phone_Number'].apply(lambda x: x[0:3]+'-'+x[3:6]+'-'+x[6:10])
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace('nan--','')
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace('Na--','')
```

```python
# Expanding the Address column

df[['Street_Address','State','Zip_Code']]=df['Address'].str.split(',',n=2,expand=True)
```

```python
df=df.drop(columns='Address')
```

```python
# replace the yes/no
```

```python
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('Yes','Y')
```

```python
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('No','N')
```

```python
df['Paying Customer']=df['Paying Customer'].str.replace('Yes','Y')
```

```python
df['Paying Customer']=df['Paying Customer'].str.replace('No','N')
```

```python
# Difference between N/a & NaN : N/a can be replaced but NaN is empty/default cannot be replaced
df=df.replace('N/a','')
```

```python
# Let's fill the empty : NaN
```

```python
df=df.fillna('')
```

```python
df
```

```python
# Loop through this col 'Do_Not_Contact' , look for Y so that we could select only who we to contact
for x in df.index:
    if df.loc[x,'Do_Not_Contact']=='Y':
        df.drop(x,inplace=True)
```

```python
df
```

```python
# remove blanks :

for x in df.index:
    if df.loc[x,'Phone_Number']=='':
        df.drop(x,inplace=True)
```

```python
df
```

```python
# lets reset the index
df.reset_index(drop=True)
```
