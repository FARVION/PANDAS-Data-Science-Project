# EDA in PANDAS

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("D:\EDA-Data-World-Population.csv")
df
```

```python
# We could see scientific notation we could change it :

pd.set_option('display.float_format',lambda x: '%.2f'%x)
```

```python
df.info()
df.describe()
df.isnull().sum()
df.nunique()
```

```python
# Strategic way to look at tip values
df.sort_values(by='2022 Population',ascending=False).head()
```

```python
# Coorelation between numeric values :

df.corr(numeric_only = True)
```

```python
# Visulaize this :
sns.heatmap(df.corr(numeric_only = True),annot=True)
plt.rcParams['figure.figsize']=(20,7)
plt.show()
```

```python
df.groupby('Continent').mean(numeric_only = True).sort_values(by='2022 Population',ascending=False)
```
